# Now that we've Run our Analysis on Equality of Mobility, let's look at some Universities that really stand out across a variety of metrics
We'll start by importing SQL database so that we can do some Querying.  We're putting our Querying skills to the test!

```python
import pandas as pd
from pandas.io import sql
import sqlite3
```

```python
#Connect the database to a SQLite database that can be saved in single file
#These databases are stored in a very efficient manner and allow fast querying
conn = sqlite3.connect('dat-test.db')

# Data is moved to the database through the `to_sql` command, similar to the `to_csv` command
um_table = pd.read_csv(r"C:\Users\Owner\Desktop\GA\FINAL PROJECT\Equality of Upward Mobility Project.csv", low_memory=False)
um_table.head(5)
```

```python
um_table.to_sql('um_table',
            con=conn,
            if_exists='replace',
            index=False)

sql.read_sql('select * from um_table limit 5', con=conn)
```

```python
#Let's find schools that have High Upward Mobility and Low Social Eliteness (ie. Their success not directly attributable to Social Elite environment)

sql.read_sql("""SELECT name_x, social_eliteness_score, upward_mobility_score, sticker_price_preds, bang4buck
                FROM um_table
                WHERE upward_mobility_score > 1 and social_eliteness_score < 0.9 and upward_mobility_bucket = "Best"
                ORDER BY upward_mobility_score desc
                LIMIT 50""", con=conn)
```

```python
#Let's Look at the Best Overall Bang for the Buck Schools with Minimum Upward Mobility Score

sql.read_sql("""SELECT name_x, tier_x, upward_mobility_score, sticker_price_preds, bang4buck
                FROM um_table
                WHERE upward_mobility_score > 1 and bang4buck > 1
                ORDER BY bang4buck desc
                LIMIT 30""", con=conn)

#Note: Tier 5 schools are "Selective Public", Tier 3 Schools are "Highly Selective Public"
```

```python
#Let's Look at the Best Overall Bang for the Buck Schools

sql.read_sql("""SELECT name_x, tier_x, upward_mobility_score, sticker_price_preds, bang4buck
                FROM um_table
                WHERE bang4buck > 2
                ORDER BY bang4buck desc
                LIMIT 30""", con=conn)

#Lots of Community Colleges here at cheap prices (but doesn't account the Cost of Living)
```

```python
#Let's find scools that have High Upward Mobility and Low Sticker Price (ie. Very affordable to attend)

sql.read_sql("""SELECT name_x, sticker_price_preds, upward_mobility_score, upward_mobility_bucket
                FROM um_table
                WHERE upward_mobility_score > 1 and sticker_price_preds < 10000 and upward_mobility_bucket = "Best"
                ORDER BY upward_mobility_score desc
                LIMIT 50""", con=conn)
```

```python
#Let's do that same study, but look for best 'bang for the buck' schools instead

sql.read_sql("""SELECT name_x, bang4buck, upward_mobility_score, tier_x, sticker_price_preds
                FROM um_table
                WHERE upward_mobility_score > 1 and bang4buck > 1.7
                ORDER BY upward_mobility_score desc
                LIMIT 100""", con=conn)

#If you can get into a Maritime Academy, that would be the best option!
#outside of that, there's many tech school options
#California Polytechnic does a great job
#The California & NY State University system of New York are well-represented
```

```python
#Let's do the opposite study, but look for worst 'bang for the buck' schools and low mobility scores instead

sql.read_sql("""SELECT name_x, bang4buck, upward_mobility_score, tier_x, sticker_price_preds
                FROM um_table
                WHERE upward_mobility_score < 0.6 and bang4buck < 0.3
                ORDER BY bang4buck asc
                LIMIT 100""", con=conn)

#Many of these are Less than 2-year schools (Tier 12) charging exorbitant rates
#Many "Beauty" and "Arts" schools atop this list
#Most of these schools are Selective Private Schools (Tier_6).  Stay away from these!
#Many are Church-affiliated schools, perhaps charging a large sticker price to subsidize religious expenses
#I would expect many of these Universities to go out of business if students realize they're not gaining upward mobility from large tuition payments
```

```python
#Check out good bang 4 buck and socially elite schools
sql.read_sql("""SELECT name_x, bang4buck, upward_mobility_score, sticker_price_preds, social_eliteness_score, tier_x
                FROM um_table
                WHERE bang4buck > 1.5 and social_eliteness_score > 1
                ORDER BY upward_mobility_score desc
                LIMIT 50""", con=conn)
```

```python
#Let's look at relative selectivity and find schools that do a good job despite being non-selective (anybody can gain entry)

sql.read_sql("""SELECT name_x, tier_x, bang4buck, upward_mobility_score, sticker_price_preds, rejection_rate_preds
                FROM um_table
                WHERE upward_mobility_score > 1 and rejection_rate_preds < 0.2 and bang4buck > 1
                ORDER BY upward_mobility_score desc
                LIMIT 50""", con=conn)

#Many of these schools are Tier 5:  Selective Public Scools
```

```python
#Who are the Most Selective Schools and how do they do?

sql.read_sql("""SELECT name_x, barrons, bang4buck, upward_mobility_score, sticker_price_preds, rejection_rate_preds, social_eliteness_score
                FROM um_table
                WHERE rejection_rate_preds > 0.85
                ORDER BY rejection_rate_preds desc
                LIMIT 50""", con=conn)

#Not surprisingly, the most selective schools have very high rejection rates & social eliteness schools as well as mobility rates
#most of these schools are Tier 1
```

```python
#Find the Cheapest Schools with the Most Mobility under $7000 in tuition

sql.read_sql("""SELECT name_x, sticker_price_preds, upward_mobility_score, bang4buck
                FROM um_table
                WHERE sticker_price_preds < 7000 and upward_mobility_score > 1
                ORDER BY upward_mobility_score desc
                LIMIT 50""", con=conn)

#California State Universities are where you want to be if you want cheap, quality in-state education!
#Perhaps this is why their taxes are so high - to fund higher education
#One caveate is that Cost of Living Adjustment is not factored in - Should I account for this in the Upward Mobility Score?
```

```python
#Let's Check out 2-year Public Schools (Tier 9) that stand out

sql.read_sql("""SELECT name_x, tier_x, sticker_price_preds, upward_mobility_score, upward_mobility_bucket, social_eliteness_score, bang4buck
                FROM um_table
                WHERE (tier_x = 9 or tier_x = 11) and upward_mobility_score > 0.9
                ORDER BY upward_mobility_score desc
                LIMIT 50""", con=conn)

#Here, we're looking at both 2-year Public & Private profit and non-for-profit
#Many of the schools are technical colleges, preparing students for the real world
```

```python
#Let's Check schools with High SAT Average versus Mobility Score

sql.read_sql("""SELECT name_x, sat_avg_2013, upward_mobility_score, upward_mobility_bucket, social_eliteness_score
                FROM um_table
                WHERE sat_avg_2013 > 1400
                ORDER BY sat_avg_2013 desc
                LIMIT 50""", con=conn)

#All schools with Average SAT Scores higher than 1400 have excellent Mobility Rates as well as Social Eliteness
#Perhaps this indicates Social Eliteness impacts SAT Scores, which the College Board is now taking action on, including a 3rd score for social background
#Slight outliers of Vassar, Williams & Haverford, which are more Liberal Arts oriented
```

```python
#Finally, drum roll please.....let's look at the overall, absolute BEST schools for Upward Mobility!

sql.read_sql("""SELECT name_x, tier_x, upward_mobility_score, rejection_rate_preds, sticker_price_preds, bang4buck
                FROM um_table
                WHERE upward_mobility_score > 1.5
                ORDER BY upward_mobility_score desc
                LIMIT 50""", con=conn)
```
